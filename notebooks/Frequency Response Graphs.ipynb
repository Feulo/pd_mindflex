```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

N=6
b_b, a_b = sig.butter(N,[12.5/256,30/256],output='ba',btype='bandpass')
b_c1, a_c1 = sig.cheby1(N,1,[12.5/256,30/256],output='ba',btype='bandpass')
b_c2, a_c2 = sig.cheby2(N,50,[12.5/256,30/256],output='ba',btype='bandpass')
b_e, a_e = sig.ellip(N,0.25,50,[12.5/256,30/256],output='ba',btype='bandpass')

w_b, h_b = sig.freqz(b_b,a_b)
w_c1, h_c1 = sig.freqz(b_c1,a_c1)
w_c2, h_c2 = sig.freqz(b_c2,a_c2)
w_e, h_e = sig.freqz(b_e,a_e)

plt.plot(w_b*256/np.pi,20*np.log10(abs(h_b)),color='purple',label='Butterworth')
plt.plot(w_c1*256/np.pi,20*np.log10(abs(h_c1)),color='green',label='Chebyshev 1')
plt.plot(w_c2*256/np.pi,20*np.log10(abs(h_c2)),color='blue',label='Chebyshev 2')
plt.plot(w_e*256/np.pi,20*np.log10(abs(h_e)), color='red',label='Cauer')
plt.axis([5,40,-20,2])
plt.axvline(30, color='black',ls='--', label='Frequncia de corte') # cutoff frequency
plt.axvline(12.5, color='black',ls='--') # cutoff frequency
plt.xlabel('Frequncia [Hertz]')
plt.ylabel('Amplitude [dB]')
plt.title('Comparao da resposta em frequncia dos filtros IIR passa-banda')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid()
plt.savefig('iir_resp.png', bbox_inches='tight')
plt.show()
```

```python
b_d, a_d = sig.ellip(3,0.25,50,[0.5/256,2.75/256],output='ba',btype='bandpass')
b_t, a_t = sig.ellip(4,0.25,50,[3.5/256,6.75/256],output='ba',btype='bandpass')
b_al, a_al = sig.ellip(4,0.25,50,[7.5/256,9.25/256],output='ba',btype='bandpass')
b_ah, a_ah = sig.ellip(4,0.25,50,[10/256,11.75/256],output='ba',btype='bandpass')
b_bl, a_bl = sig.ellip(5,0.25,50,[13/256,16.75/256],output='ba',btype='bandpass')
b_bh, a_bh = sig.ellip(5,0.25,50,[18/256,29.75/256],output='ba',btype='bandpass')
b_gl, a_gl = sig.ellip(5,0.25,50,[31/256,39.75/256],output='ba',btype='bandpass')
b_gm, a_gm = sig.ellip(5,0.25,50,[41/256,49.75/256],output='ba',btype='bandpass')

w_d, h_d = sig.freqz(b_d,a_d)
w_t, h_t = sig.freqz(b_t,a_t)
w_al, h_al = sig.freqz(b_al,a_al)
w_ah, h_ah = sig.freqz(b_ah,a_ah)
w_bl, h_bl = sig.freqz(b_bl,a_bl)
w_bh, h_bh = sig.freqz(b_bh,a_bh)
w_gl, h_gl = sig.freqz(b_gl,a_gl)
w_gm, h_gm = sig.freqz(b_gm,a_gm)

plt.plot(w_d*256/np.pi,20*np.log10(abs(h_d)),label='Delta')# 0.5 - 2.75 Hz")
plt.plot(w_t*256/np.pi,20*np.log10(abs(h_t)),label='Theta')# 3.5 - 6.75 Hz')
plt.plot(w_al*256/np.pi,20*np.log10(abs(h_al)),label='Low Alpha')# 7.5 - 9.25 Hz')
plt.plot(w_ah*256/np.pi,20*np.log10(abs(h_ah)),label='High Alfa')# 10 - 11.75 Hz')
plt.plot(w_bl*256/np.pi,20*np.log10(abs(h_bl)),label='Low Beta')# 13 - 16.75 Hz')
plt.plot(w_bh*256/np.pi,20*np.log10(abs(h_bh)),label='High Beta')# 18 - 29.75 Hz')
plt.plot(w_gl*256/np.pi,20*np.log10(abs(h_gl)),label='Low Gamma')# 31 - 39.75 Hz')
plt.plot(w_gm*256/np.pi,20*np.log10(abs(h_gm)),label='Mid Gamma')# 41 - 49.75 Hz')
plt.xlabel('Frequency [Hertz]')
plt.ylabel('Amplitude [dB]')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Bandpass Filters frequency response for diferent EEG bands')
plt.axis([0,60,-30,3])
plt.grid()
#graph = plt.figure()
plt.savefig('freq_res.png', bbox_inches='tight')
plt.show()
```

```python
b_a, a_a = sig.ellip(5,0.25,50,[7.5/256,11.75/256],output='ba',btype='bandpass')
b_b, a_b = sig.ellip(6,0.25,50,[13/256,29.75/256],output='ba',btype='bandpass')

w_a, h_a = sig.freqz(b_a,a_a)
w_b, h_b = sig.freqz(b_b,a_b)

plt.plot(w_a*256/np.pi,20*np.log10(abs(h_a)),label='Alpha 8 - 13 Hz')
plt.plot(w_al*256/np.pi,20*np.log10(abs(h_al)),label='Low Alpha 7.5 - 9.25 Hz')
plt.plot(w_ah*256/np.pi,20*np.log10(abs(h_ah)),label='High Alfa 10 - 11.75 Hz')
plt.xlabel('Frequency [Hertz]')
plt.ylabel('Amplitude [dB]')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Comparison beetween the whole alpha band filter and the low alpha and high alpha filters')
plt.axis([5,15,-30,3])
plt.grid()
plt.show()
```

```python
plt.plot(w_b*256/np.pi,20*np.log10(abs(h_b)),label='Beta 13 - 30 Hz')
plt.plot(w_bl*256/np.pi,20*np.log10(abs(h_bl)),label='Low Beta 13 - 16.75 Hz')
plt.plot(w_bh*256/np.pi,20*np.log10(abs(h_bh)),label='High Beta 18 - 29.75 Hz')
plt.xlabel('Frequency [Hertz]')
plt.ylabel('Amplitude [dB]')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Comparison beetween the whole beta band filter and the low beta and high beta filters')
plt.axis([12,32,-30,3])
plt.grid()
plt.show()
```
