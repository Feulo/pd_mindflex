```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import spline
points=2500
teste = "teste9.txt"
mydir = "/home/feulo/pd-externals/mindflex/teste_fisica_raw/"
x = np.loadtxt(mydir+ teste)/2**11
X = np.fft.rfft(x, n=points,norm="ortho")
y = 10*np.log10(np.absolute(X))
x_fft = np.fft.rfftfreq(points,1.0/512)
plt.plot(x_fft,y,label='Curva obtida')
plt.axis([0,256,-30, 0])
#plt.axhline(-5.1, color='red',label='Referência')
plt.ylabel('Magnitude [db]')
plt.title('Curva de resposta em frequência da interface Mindflex')
#lgd = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.savefig("espectro_sweep_lin_linear",bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.savefig("espectro_sweep_lin_linear")
plt.show()
```
