# Testes com a interface Mindflex

Pequeno relatório dos experimentos conduzidos no labóratório de sensores do Instituto de Física da USP

## Equipamentos e Setup

Para induzir o sinal elétrico no sensor da interface foi o utilizado o gerador de função **HP 33120A Function / Arbitrary Waveform Generator 15 MHz**

<img src="images/gerador.jpg" width="600">

Como a menor amplitude pico a pico que o gerador consegue atingir é de 50mV, foi usado um divisor de tensão feitos com dois resistores, na proporção 1/1000.

<img src="images/divisor2.jpg" width="600">

A saídas do divisor de tensão é então conectada ao sensor e as referências da interface Mindflex.

<img src="images/DSC_0011.JPG" width="600">


A relação de etrada e saída real do divisor de tensão é medida inserindo um sinal senoidal com amplitude conhecida e medindo a amplitude pico a pico da saída do divisor

$$ R_{io} = V_o/V_i = 16.10 mV / 20.1 V =  0.800995025 \times 10^{-3}$$

Onde a amplitude pico a pico da saída foi obitida através da média de 8 instantes. A realização da média foi necessária devido ao ruído do ambiente e a precisão do osciloscópio.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(512)
x= 1.39*16.1/20.1*np.sin(2*np.pi*4/512*t)
x_d = 80*10**-3*np.sin(2*np.pi*2/512*t) # de 70 a 100 microvolts
x_t = 50*10**-3*np.sin(2*np.pi*5/512*t) #50 micro volts
x_a = 50*10**-3*np.sin(2*np.pi*10/512*t) #de 20 a 60 micrvolts
x_b = 13*10**-3*np.sin(2*np.pi*15/512*t)    #de 10 a 15 microvolts
x_g = 2*10**-3*np.sin(2*np.pi*40/512*t)
plt.plot(t/512,x,label='Reference')
plt.plot(t/512,x_d, label='Delta')
plt.plot(t/512,x_t, label='Theta')
plt.plot(t/512,x_a, label='Alpha')
plt.plot(t/512,x_b, label='Beta')
plt.plot(t/512,x_g, label='Gamma')
plt.axis([0,1,-1.39*16.1/20.1, 1.39*16.1/20.1])
#plt.axis([0,1,-80*10**-3, 80*10**-3])
plt.xlabel('Time [seconds]')
plt.ylabel('Amplitude [milivolts]')
plt.title('Different EEG bands Amplitude Comparision')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
