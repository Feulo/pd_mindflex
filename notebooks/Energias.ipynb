```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
af = np.loadtxt("/home/feulo/energy_alpha_filtered.txt")
bf = np.loadtxt("/home/feulo/energy_beta_filtered.txt")
laf = np.loadtxt("/home/feulo/energy_low_alpha_filtered.txt")
lai = np.loadtxt("/home/feulo/energy_low_alpha_interface.txt")
haf = np.loadtxt("/home/feulo/energy_high_alpha_filtered.txt")
hai = np.loadtxt("/home/feulo/energy_high_alpha_interface.txt")
hbf = np.loadtxt("/home/feulo/energy_high_beta_filtered.txt")
hbi = np.loadtxt("/home/feulo/energy_high_beta_interface.txt")
lbf = np.loadtxt("/home/feulo/energy_low_beta_filtered.txt")
lbi = np.loadtxt("/home/feulo/energy_low_beta_interface.txt")  
att =  np.loadtxt("/home/feulo/attention_interface.txt") 
```

```python
plt.plot(af/max(abs(af))*100 ,label='Alpha')
plt.plot(bf/max(abs(bf))*100, label='Beta')
plt.plot(att,label='Attention')
plt.axis([0,60,0,100])
plt.xlabel('Time [seconds]')
plt.ylabel('Attention')
plt.title('Mindflex Attetion X Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.plot(af,label='Alpha Filtered')
plt.plot(laf+haf, label='Soma filtered')
plt.plot(lai+hai,label='Soma Interface')
#plt.axis([0,60,0,100])
plt.xlabel('Time [seconds]')
plt.ylabel('Attention')
plt.title('Mindflex Attetion X Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.plot(bf,label='Beta Filtered')
plt.plot(lbf+hbf, label='Soma filtered')
plt.plot(hbi+hbf,label='Soma Interface')
#plt.axis([0,60,0,100])
plt.xlabel('Time [seconds]')
plt.ylabel('Attention')
plt.title('')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.plot(af/bf/max(af/bf)*100,label= 'af/bf')
plt.plot(bf/af/max(bf/af)*100, label='bf/af')
plt.plot(att,label='Soma Interface')
#plt.axis([0,60,0,100])
plt.xlabel('Time [seconds]')
plt.ylabel('Attention')
plt.title('')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
